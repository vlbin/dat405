## 2a. discuss why we chose this data and motivate our choice of visualization

**What data did we choose?**  
We chose a dataset from Our World in Data representing the:
> Share of persons of working age who were engaged in any activity to produce goods or provide services for pay or
profit in the agriculture sector (agriculture, hunting, forestry and fishing). [Source](https://ourworldindata.org/grapher/share-of-the-labor-force-employed-in-agriculture)

We also chose a dataset from the same source representing life satisfaction and the life expectancy dataset provided in
the course material to see if any correlation could be found between these factors.

Our World in Data defines the data surrounding life satisfaction as follows:
> Life satisfaction is self-reported as the answer to the following question: “Please imagine a ladder, with steps
numbered from 0 at the bottom to 10 at the top. The top of the ladder represents the best possible life for you and the
bottom of the ladder represents the worst possible life for you. On which step of the ladder would you say you
personally feel you stand at this time?” [Source](https://ourworldindata.org/grapher/happiness-cantril-ladder?time=2015)

**Questions to answer from this data:**
* Does the percentage of population in agriculture correlate in anyway with the life satisfaction of the population?

    * If so, how can this be explained?

**Why did we choose this type of visualization for the data?**  
The scatter plots gives us a general oversight of if life satisfaction/expectancy and agriculture correlate, and if so, how.  
We used tables to get the outliers from these sets so we can identify them and try to analyze why these countries are not following the trend. And hopefully see 
what other factors are affecting them.


```python
import pandas as pd


#Load the datasets into pandas DataFrames
life_satisfaction = pd.read_csv('./life_satisfaction.csv')
agriculture = pd.read_csv('./agriculture_employment.csv')
life_exp = pd.read_csv('./life-expectancy.csv')
gdp = pd.read_csv('./GDP-2015.csv')

separator = '------------------------------------------------------------'

#Select life expectancy data from year 2015 and merge with gdp dataset
life_satisfaction = life_satisfaction[life_satisfaction['Year'] == 2015]
life_satisfaction = life_satisfaction.rename(columns = {list(life_satisfaction)[-1]: 'Satisfaction'})

life_exp = life_exp[life_exp['Year'] == 2015]
life_exp = life_exp.rename(columns = {list(life_exp)[-1]: 'Life expectancy'})

agriculture = agriculture[agriculture['Year'] == 2015]
agriculture = agriculture.rename(columns = {list(agriculture)[-1]: 'Agriculture'})

agr_sat = pd.merge(life_satisfaction, agriculture, on=['Code', 'Entity', 'Year'])
combined = pd.merge(agr_sat, life_exp, on=['Code', 'Entity', 'Year'])

sat_mean = combined['Satisfaction'].mean()
agr_mean = combined['Agriculture'].mean()
exp_mean = combined['Life expectancy'].mean()

high_agr = combined[combined['Agriculture'] > agr_mean]
low_agr = combined[combined['Agriculture'] < agr_mean]

low_agr_low_sat = low_agr[low_agr['Satisfaction'] < sat_mean]
high_agr_high_sat = high_agr[high_agr['Satisfaction'] > sat_mean]

low_agr_low_exp = low_agr[low_agr['Life expectancy'] < exp_mean]
high_agr_high_exp = high_agr[high_agr['Life expectancy'] > exp_mean]
```

**Plot: Life *satisfaction* as a function of agriculture employment share**

```python
combined.plot.scatter(x = 'Agriculture', y='Satisfaction', c='DarkBlue')
```

**Plot: Life *expectancy* as a function of agriculture employment share**

```python
combined.plot.scatter(x = 'Agriculture', y='Life expectancy', c='DarkBlue')
```
