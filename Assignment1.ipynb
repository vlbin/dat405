```python
import pandas as pd

#Load the datasets into pandas DataFrames
life_exp = pd.read_csv('./life-expectancy.csv')
gdp = pd.read_csv('./GDP-2015.csv')

#Select life expectancy data from year 2015 and merge with gdp dataset
common = life_exp[life_exp['Year'] == 2015].merge(gdp, on=['Code', 'Entity', 'Year'])

mean = common['Life expectancy (years)'].mean()
std = common['Life expectancy (years)'].std()

higher = common[common['Life expectancy (years)'] > mean + std]['Entity']

print(higher)
```

```python
import pandas as pd

#Load the datasets into pandas DataFrames
life_exp = pd.read_csv('./life-expectancy.csv')
gdp = pd.read_csv('./GDP-2015.csv')

#Select life expectancy data from year 2015 and merge with gdp dataset
common = life_exp[life_exp['Year'] == 2015].merge(gdp, on=['Code', 'Entity', 'Year'])

gdp_mean = common['GDP per capita'].mean()
le_mean = common['Life expectancy (years)'].mean()

high_gdp = common[common['GDP per capita'] > gdp_mean]
low_le = high_gdp[high_gdp['Life expectancy (years)'] < le_mean]

print(low_le)
```
